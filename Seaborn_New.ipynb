```python
import seaborn as sns
import matplotlib.pyplot as plt

propinas = sns.load_dataset("tips")
print(propinas)

sns.relplot(data=propinas,
            x="total_bill",
            y="tip"),
```

```python
sns.relplot(data=propinas,
            x="total_bill",
            y="tip", 
            kind="line")
```

```python
sns.relplot(data=propinas,
            x="total_bill",
            y="tip", 
            kind="line",
            hue="smoker",
            style="smoker",
            size="size")
```

```python
sns.relplot(data=propinas,
            x="total_bill",
            y="tip", 
            hue="smoker",
            style="smoker",
            size="size")
```

```python
sns.relplot(data=propinas,
            x="total_bill",
            y="tip", 
            hue="smoker",
            style="smoker",
            size="size",
            col="time")
```

```python
sns.relplot(data=propinas,
            x="total_bill",
            y="tip", 
            hue="smoker",
            style="smoker",
            size="size",
            col="time",
            legend=None)
```
