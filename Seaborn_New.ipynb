```python
import seaborn as sns
import matplotlib.pyplot as plt

pinguinos = sns.load_dataset("penguins")


sns.histplot(data=pinguinos,
            x="flipper_length_mm",
            hue="species",
            multiple="stack")
plt.show()
```

```python
sns.displot(data=pinguinos,
            x="flipper_length_mm",
            hue="species",
            multiple="stack")
plt.show()
```

```python
sns.displot(data=pinguinos,
            x="flipper_length_mm",
            hue="species",
            multiple="stack",
            col="species");
plt.show()
```

```python

sns.relplot(data=pinguinos,
            x="flipper_length_mm",
            y="bill_length_mm",
            col="sex")
plt.show()
```

```python

graf = sns.relplot(data=pinguinos,
            x="flipper_length_mm",
            y="bill_length_mm",
            col="sex")
graf.set_axis_labels("Largo Aletas (mm)")
plt.show()
```
